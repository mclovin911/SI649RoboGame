```python
import altair as alt
```

```python
import pandas as pd
```

```python
ex1 = pd.read_csv('./robogame-v0.5/server/example1/examplematch1.robotdata.csv')
```

```python
vis_data = ex1.iloc[:100, -11:]
```

```python
vis_data
```

```python
alt.Chart(vis_data).mark_point().encode(
    alt.X('Sonoreceptors:Q'),
    alt.Y('Cranial Uplink Bandwidth:Q'),
    alt.Color('Productivity:Q')
) + alt.Chart(vis_data).mark_point(color="red").encode(
    alt.X('Sonoreceptors:Q'),
    alt.Y('Cranial Uplink Bandwidth:Q'),
).transform_filter(
    alt.FieldRangePredicate(field='Productivity', range=[-20, 20])
)
```
